```python
#tokenize and lemmatize code found: https://www.machinelearningplus.com/nlp/lemmatization-examples-python/

import nltk
nltk.download('wordnet')
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
#Adam used the nested code here, but I ~think~ I've done the same thing here but in more steps
mcarthur_tokens = nltk.sent_tokenize(tags_str)
```

```python
lemmatizer = WordNetLemmatizer()

macarthur_lemmatize = ' '.join([lemmatizer.lemmatize(m) for m in mcarthur_tokens])
```

```python
import gensim
```

```python
dictionary = gensim.corpora.Dictionary(macarthur_lemmatize)
corpus = [dictionary.doc2bow(l) for l in macarthur_lemmatize]
model = gensim.models.LdaMulticore(corpus, id2word=dictionary, num_topics=10, alpha = 0.9, eta=0.9)
```
